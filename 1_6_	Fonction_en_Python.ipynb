# **Exemple d’application 1.21**

```python
def addition(a, b):
    resultat = a + b
    return resultat

# Saisie des valeurs par l'utilisateur
a = float(input("Entrez la première valeur de a : "))
b = float(input("Entrez la deuxième valeur de b: "))

# Appel de la fonction
x = addition(a, b)
print(f"Le résultat de {a} + {b} = {x}")
```

# **Exemple d’application 1.22**

```python
import pandas as pd
import numpy as np

df = pd.read_excel("/content/Exemple_1.3.xlsx")

def calculer_ctr(df):
    """Calcule le CTR"""
    return (df['clics'] / df.get('impressions', df['visites'])) * 100

def calculer_cpa(df):
    """Calcule le CPA"""
    return df['budget'] / df['achats']

def calculer_roi(df):
    """Calcule le ROI"""
    return ((df['ca'] - df['budget']) / df['budget']) * 100

def analyser_marketing(df):
    """Fonction principale simplifiée"""
    # Calculer avec les fonctions
    df['CTR (%)'] = calculer_ctr(df)
    df['CPA (€)'] = calculer_cpa(df)
    df['ROI (%)'] = calculer_roi(df)

    # Afficher
    print("Indicateurs calculés avec fonctions :")
    print(df[['CTR (%)', 'CPA (€)', 'ROI (%)']].head())

    return df
# Appel de la fonction
df_resultat = analyser_marketing(df)
```

# **Exemple d’application 1.13**

```python
# Création dynamique d'une liste de clients
clients_saisis = []
while True:
    nom = input("Nom du client (ou 'stop' pour terminer) : ")
    if nom.lower() == 'stop':
        break
    email = input(f"Email de {nom} : ")
    clients_saisis.append({"nom": nom, "email": email})
    print(f"Client ajouté ! Total : {len(clients_saisis)}\n")
print(f"\n Liste finale : {len(clients_saisis)} clients")
for client in clients_saisis:
    print(f"• {client['nom']} - {client['email']}")
```

# **Exemple d’application 1.14 **

```python
print("SUIVI DES ACHATS CLIENTS")
print("=" * 45)
# TUPLES : Noms des clients (immuables)
clients_tuples = (
    ("Alice", "Martin"),
    ("Bob", "Dupont"),
    ("Charlie", "Wilson"),
    ("Diana", "Roy")
)
# LISTE : Montants des achats (modifiables)
montants_achats = [150.00, 85.50, 45.00, 210.75]
print("LISTE DES CLIENTS :")
for i, client in enumerate(clients_tuples):
    prenom, nom = client
    print(f"   {i+1}. {prenom} {nom}")
print("\n" + "=" * 45)
print("MONTANTS ACTUELS DES ACHATS :")
for i, montant in enumerate(montants_achats):
    prenom, nom = clients_tuples[i]
    print(f"   {prenom} {nom} : {montant}€")
print("\n" + "=" * 45)
print("MODIFICATION DES MONTANTS")
print("=" * 45)
# L'utilisateur modifie les montants
for i in range(len(montants_achats)):
    prenom, nom = clients_tuples[i]
    ancien_montant = montants_achats[i]
    nouveau_montant = float(input(f"Montant pour {prenom} {nom} (actuel: {ancien_montant}€) : "))
    # MODIFICATION DE LA LISTE
    montants_achats[i] = nouveau_montant
    print(f"Mis à jour : {ancien_montant}€ → {nouveau_montant}€")
print("\n" + "=" * 45)
print("RÉCAPITULATIF")
print("=" * 45)
total_ventes = 0
for i, client in enumerate(clients_tuples):
    prenom, nom = client
    montant = montants_achats[i]
    total_ventes += montant
    print(f"{prenom} {nom} : {montant}€")
print("-" * 45)
print(f"TOTAL DES VENTES : {total_ventes}€")
print("=" * 45)
# DÉMONSTRATION
print("\n DÉMONSTRATION :")
print(" Liste (montants) modifiable :")
montants_achats[1] = 100.00  #  Possible
print(f"   Bob : {montants_achats[1]}€")

print("\n Tuple (noms) non modifiable :")
try:
    clients_tuples[1] = ("Éric", "Durand")  #  Impossible
except TypeError:
    print("  Protection des noms activée")
```

# **Exemple d’application 1.15**

```python
# Liste de 3 clients
clients = [
    {
        "nom": "Dupont",
        "prenom": "Marie",
        "email": "marie.dupont@email.com",
        "adresse": {
            "rue": "12 Rue de Paris",
            "ville": "Lyon",
            "code_postal": 69001
        },
        "montant_achats": 45.98  # 25.99 + 19.99
    },
    {
        "nom": "Martin",
        "prenom": "Pierre",
        "email": "pierre.martin@email.com",
        "adresse": {
            "rue": "5 Avenue des Champs",
            "ville": "Paris",
            "code_postal": 75008
        },
        "montant_achats": 120.50
    },
    {
        "nom": "Bernard",
        "prenom": "Sophie",
        "email": "sophie.bernard@email.com",
        "adresse": {
            "rue": "8 Boulevard de la Liberté",
            "ville": "Marseille",
            "code_postal": 13001
        },
        "montant_achats": 89.75
    }
]
# Demander à l'utilisateur de saisir un nom
nom_recherche = input("Entrez le nom du client à rechercher : ")
# Rechercher le client
client_trouve = None
for client in clients:
    if client["nom"].lower() == nom_recherche.lower():
        client_trouve = client
        break

# Afficher les informations si le client est trouvé
if client_trouve:
    print(f"\n Informations du client {client_trouve['nom']} :")
    print("=" * 40)
    print(f" Nom complet : {client_trouve['prenom']} {client_trouve['nom']}")
    print(f" Email : {client_trouve['email']}")
    print(f" Adresse : {client_trouve['adresse']['rue']}")
    print(f"   {client_trouve['adresse']['ville']} {client_trouve['adresse']['code_postal']}")
    print(f"Montant total des achats : {client_trouve['montant_achats']} €")
    print("=" * 40)
else:
    print(f"\n Aucun client trouvé avec le nom '{nom_recherche}'")
```

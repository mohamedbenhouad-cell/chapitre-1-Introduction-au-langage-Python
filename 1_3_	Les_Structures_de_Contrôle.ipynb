# **Exemple d’application 1.6 **

```python
# Donnée fixe - montant des achats du client
montant_achats = 100  # €
# Structure conditionnelle simple
if montant_achats >= 1000:
    # Livraison gratuite si ≥ 1000€
    print(" LIVRAISON GRATUITE")
    print(f"Total : {montant_achats}€")
else:
      # Frais de livraison de 15 € si < 1000€
    print("LIVRAISON STANDARD")
    print(f"Total : {montant_achats + 15.00:.2f}€")
```

# **Exemple d’application 1.7 **

```python
# Donnée fixe - montant des achats du client
montant_achats = 100  # €
# Structure avec 3 conditions
if montant_achats >= 1000:
    # Condition 1 : ≥ 1000€ → Livraison gratuite + cadeau
    print("LIVRAISON GRATUITE + CADEAU")
    print(f"Total : {montant_achats}€")
elif montant_achats >= 500:
    # Condition 2 : ≥ 500€ → Livraison gratuite
    print("LIVRAISON GRATUITE")
    print(f"Total : {montant_achats}€")
else:
    # Condition 3 : < 500€ → Frais de livraison
    frais_livraison = 15.00
    total = montant_achats + frais_livraison
    print("LIVRAISON STANDARD")
    print(f"Frais livraison : {frais_livraison}€")
    print(f"Total : {total:.2f}€")
    # Calcul pour la gratuité
    manquant = 500 - montant_achats
    print(f" Ajoutez {manquant:.2f}€ pour la livraison gratuite !")
```

# **Exemple d’application 1.8**

```python
# Données du client
montant_depense_annuel = 850.00
nombre_achats = 8
anciennete_mois = 18
# Structure conditionnelle de segmentation
if montant_depense_annuel >= 1000 and nombre_achats >= 10:
    segment = "VIP"
    promotion = 25
    message = "Offre exclusive VIP !"
elif montant_depense_annuel >= 500 or nombre_achats >= 5:
    segment = "Fidèle"
    promotion = 15
    message = "Merci pour votre fidélité !"
elif anciennete_mois < 3:
    segment = "Nouveau"
    promotion = 10
    message = "Bienvenue ! Profitez de votre offre de bienvenue"
else:
    segment = "Standard"
    promotion = 5
    message = "Découvrez nos promotions"
# Affichage du résultat
print(f"Dépense annuelle : {montant_depense_annuel}€")
print(f"Nombre d'achats : {nombre_achats}")
print(f"Ancienneté : {anciennete_mois} mois")
print("-" * 40)
print(f"Segment attribué : {segment}")
print(f"Promotion offerte : {promotion}%")
```

# **Exemple d’application 1.10 **

```python
# Liste des clients à contacter
clients = ["alice@email.com", "bob@entreprise.com", "charlie@mail.com", "diana@pro.com"]
print("CAMPAGNE EMAIL AUTOMATISÉE")
print("=" * 35)
# Boucle for simple pour traiter chaque client
for email in clients:
    # Ce code s'exécute pour CHAQUE email dans la liste
    print(f"Email envoyé à : {email}")
    print(f"  Sujet : Nouvelle promotion spéciale")
    print(f"   Statut : Envoyé avec succès")
    print("-" * 20)
```

# **Exemple  d’application 1.11 **

```python
# Objectif de vente quotidien
budget_quotidien = 5000
ventes_cumulees = 0
numero_vente = 1
print(f"Objectif quotidien : {budget_quotidien}€")
print("-" * 40)
# Boucle while : continue tant que le budget n'est pas atteint
while ventes_cumulees < budget_quotidien:
    # Simulation d'une vente
    montant_vente = float(input(f"Vente {numero_vente} - Montant (€) : "))
    ventes_cumulees += montant_vente
    numero_vente += 1
    print(f" Ventes cumulées : {ventes_cumulees}€")
    print(f" Reste à atteindre : {budget_quotidien - ventes_cumulees}€")
    print("-" * 30)
# Résultat final
print(f"OBJECTIF ATTEINT !")
print(f"Total des ventes : {ventes_cumulees}€")
print(f"Nombre de ventes : {numero_vente - 1}")
print("=" * 40)
```

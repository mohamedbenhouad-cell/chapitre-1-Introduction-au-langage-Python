# **Exemple d’application 1.5**

```python
# === DONNÉES DU CLIENT ===
nom_client = "Marie Dubois"
total_achats = 8
montant_total_depense = 850.00
dernier_achat_jours = 15
# === CALCUL DES STATUTS BOOLÉENS ===
est_vip = total_achats >= 5 and montant_total_depense >= 500
est_actif_recent = dernier_achat_jours <= 30
est_fidele = total_achats > 3
peut_recevoir_offre = est_actif_recent and not est_vip
# === AFFICHAGE DES RÉSULTATS ===
print("ANALYSE DU PROFIL CLIENT")
print("=" * 30)
print(f"Client : {nom_client}")
print(f"Achats totaux : {total_achats}")
print(f"Dépense totale : {montant_total_depense}€")
print(f"Dernier achat : il y a {dernier_achat_jours} jours")
print("-" * 35)
print(f"Client VIP : {est_vip}")
print(f"Client actif : {est_actif_recent}")
print(f"Client fidèle : {est_fidele}")
print(f"Peut recevoir offre spéciale : {peut_recevoir_offre}")
```

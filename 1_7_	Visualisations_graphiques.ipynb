# **Exemple d‚Äôapplication 1.26**

```python
# √âtape 1 : Importer les biblioth√®ques
import pandas as pd
import matplotlib.pyplot as plt

# √âtape 2 : Charger le fichier Excel
# Chemin vers le fichier Excel (√† adapter si besoin)
fichier = "/content/Exemple_1.2.xlsx"
# Lecture du fichier Excel dans un DataFrame pandas
df = pd.read_excel(fichier)

# √âtape 3 : D√©finir la liste des variables quantitatives √† analyser
colonnes_boxplot = [
    "Prix_moyen_produit_achet√©",
    "Fr√©quence_achat",
    "Temps_pass√©_site_web"
]

# √âtape 4 : Extraire uniquement les colonnes num√©riques √† partir des noms choisis
df_num = df[colonnes_boxplot].copy()

# √âtape 5 : Initialiser un dictionnaire pour stocker les statistiques de chaque variable
stats_vars = {}

# √âtape 6 : Boucle sur chaque variable de la liste pour calculer les statistiques
for col in colonnes_boxplot:
    # Extraire la s√©rie correspondant √† la colonne et supprimer les valeurs manquantes
    serie = df_num[col].dropna()
    # Calculer la moyenne
    moyenne = serie.mean()
    # Calculer la m√©diane
    mediane = serie.median()
    # Calculer le premier quartile Q1
    q1 = serie.quantile(0.25)
    # Calculer le troisi√®me quartile Q3
    q3 = serie.quantile(0.75)
    # Calculer l‚Äô√©cart type
    ecart_type = serie.std()

    # Calculer l‚Äôintervalle interquartile IQR = Q3 - Q1
    iqr = q3 - q1
    # Calculer la borne basse des valeurs non aberrantes
    borne_basse = q1 - 1.5 * iqr
    # Calculer la borne haute des valeurs non aberrantes
    borne_haute = q3 + 1.5 * iqr
    # Identifier les valeurs aberrantes (outliers)
    outliers = serie[(serie < borne_basse) | (serie > borne_haute)]

    # Stocker toutes les statistiques de la variable dans le dictionnaire
    stats_vars[col] = {
        "mean": moyenne,
        "median": mediane,
        "q1": q1,
        "q3": q3,
        "ecart_type": ecart_type,
        "nb_outliers": len(outliers)
    }

# √âtape 7 : Construction des bo√Ætes √† moustaches (boxplots)
# Cr√©er une nouvelle figure avec une taille d√©finie
plt.figure(figsize=(11, 6))

# Tracer le boxplot √† partir de df_num en affichant aussi la moyenne
df_num.boxplot(showmeans=True)

# √âtape 8 : Ajuster l‚Äôaxe x pour laisser de la place au texte descriptif
plt.xlim(0.5, len(colonnes_boxplot) + 1.0)

# √âtape 9 : Ajouter les points et les textes (moyenne, m√©diane, etc.) pour chaque variable
for i, col in enumerate(colonnes_boxplot, start=1):
    # R√©cup√©rer les statistiques de la variable courante
    stats = stats_vars[col]
    # R√©cup√©rer la moyenne
    m = stats["mean"]
    # R√©cup√©rer la m√©diane
    med = stats["median"]

    # Tracer un point pour la moyenne (losange vert)
    plt.scatter(i, m, color="green", marker="D", zorder=4, label="_nolegend_")
    # Tracer un point pour la m√©diane (carr√© bleu)
    plt.scatter(i, med, color="blue", marker="s", zorder=4, label="_nolegend_")

    # Construire le texte explicatif √† afficher √† droite de la bo√Æte
    texte = (f"Œº={m:.1f}\n"
             f"Med={med:.1f}\n"
             f"Q1={stats['q1']:.1f}\n"
             f"Q3={stats['q3']:.1f}\n"
             f"œÉ={stats['ecart_type']:.1f}\n"
             f"Outliers={stats['nb_outliers']}")
    # Placer ce texte l√©g√®rement √† droite de la bo√Æte (x = i + 0.25)
    plt.text(i + 0.25, med, texte,
             ha="left", va="center",
             fontsize=12,
             bbox=dict(boxstyle="round,pad=0.3",
                       fc="white", ec="gray", alpha=0.8))

# √âtape 10 : Mise en forme g√©n√©rale du graphique
# Ajouter un titre au graphique
plt.title("Bo√Ætes √† moustaches des variables clients")
# Ajouter un label √† l‚Äôaxe des ordonn√©es
plt.ylabel("Valeurs")
# Placer les noms des variables en x, avec une l√©g√®re rotation
plt.xticks(range(1, len(colonnes_boxplot) + 1), colonnes_boxplot, rotation=15)
# Ajouter une grille sur l‚Äôaxe vertical
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Ajuster automatiquement la mise en page pour √©viter les chevauchements
plt.tight_layout()
# Afficher le graphique
plt.show()




```

# **Exemple d‚Äôapplication 1.27 **

```python
# √âtape 1 : Importer les biblioth√®ques n√©cessaires
import pandas as pd          # pandas pour g√©rer le fichier Excel et les tableaux de donn√©es
import matplotlib.pyplot as plt   # matplotlib pour tracer le diagramme en barres

# √âtape 2 : D√©finir le chemin du fichier Excel contenant la base de donn√©es
fichier = "Exemple_1.2.xlsx"   # Nom du fichier Excel (adapter le chemin si besoin)

# √âtape 3 : Charger la base de donn√©es dans un DataFrame pandas
df = pd.read_excel(fichier)    # Lecture du fichier Excel dans un DataFrame

# √âtape 4 : Indiquer les trois colonnes cat√©gorielles que l‚Äôon souhaite comparer
colonnes_categ = [
    "Cat√©gorie_produits_achet√©s_dominants",  # Cat√©gorie du produit le plus achet√©
    "Cat√©gorie_produits_consult√©s",          # Cat√©gorie des produits consult√©s
    "Cat√©gorie_recommand√©e_achet√©e"         # Cat√©gorie des produits recommand√©s achet√©s
]

# √âtape 5 : Extraire ces trois colonnes de la base de donn√©es
df_cat = df[colonnes_categ]    # DataFrame ne contenant que les colonnes d‚Äôint√©r√™t

# √âtape 6 : Calculer la fr√©quence (nombre d‚Äôoccurrences) de chaque cat√©gorie pour chaque variable
freq_dict = {}                 # Dictionnaire qui stockera les tableaux de fr√©quences

for col in colonnes_categ:     # Boucle sur chaque colonne cat√©gorielle
    # value_counts() compte le nombre d‚Äôoccurrences de chaque cat√©gorie
    freq_dict[col] = df_cat[col].value_counts()

# √âtape 7 : Construire un DataFrame de fr√©quences √† partir du dictionnaire
# On transforme le dictionnaire freq_dict (cat√©gories ‚Üí fr√©quences) en DataFrame
freq_df = pd.DataFrame(freq_dict)
# On remplace les valeurs manquantes (NaN) par 0 pour √©viter des trous dans le graphique
freq_df = freq_df.fillna(0)

# üîπ Option : renommer les colonnes pour avoir une l√©gende plus courte
# On renomme la colonne "Cat√©gorie_produits_achet√©s_dominants" en "Achet√©s dominants"
# On renomme la colonne "Cat√©gorie_produits_consult√©s" en "Consult√©s"
# On renomme la colonne "Cat√©gorie_recommand√©e_achet√©e" en "Recommand√©s achet√©s"
freq_df = freq_df.rename(columns={
    "Cat√©gorie_produits_achet√©s_dominants": "Achet√©s dominants",
    "Cat√©gorie_produits_consult√©s": "Consult√©s",
    "Cat√©gorie_recommand√©e_achet√©e": "Recommand√©s achet√©s"
})

# √âtape 8 : Tracer le diagramme en barres group√©es
# On cr√©e une nouvelle figure graphique de taille 10 x 6
plt.figure(figsize=(10, 6))
# On trace le diagramme en barres √† partir du DataFrame freq_df
ax = freq_df.plot(
    kind="bar",          # Type de graphique : barres
    figsize=(10, 6),     # Taille de la figure (redondant mais sans cons√©quence)
    edgecolor="black"    # Bordure noire autour des barres pour une meilleure lisibilit√©
)

# On ajoute un titre au graphique
plt.title("Comparaison des cat√©gories de produits")
# On nomme l‚Äôaxe des abscisses
plt.xlabel("Cat√©gories de produits")
# On nomme l‚Äôaxe des ordonn√©es
plt.ylabel("Nombre de clients / occurrences")
# On incline les √©tiquettes des cat√©gories sur l‚Äôaxe x pour √©viter le chevauchement
plt.xticks(rotation=45, ha="right")
# On ajoute une grille horizontale (sur l‚Äôaxe y) en pointill√©s pour faciliter la lecture
plt.grid(axis="y", linestyle="--", alpha=0.5)

# üîπ D√©placer la l√©gende en dehors du graphique (√† droite)
# On place la l√©gende √† l‚Äôext√©rieur, l√©g√®rement √† droite du graphique
plt.legend(
    title="Variables",        # Titre de la l√©gende
    bbox_to_anchor=(1.02, 1), # Position de la bo√Æte de l√©gende (x, y) par rapport au graphique
    loc="upper left",         # On ancre la l√©gende par son coin sup√©rieur gauche
    borderaxespad=0.0        # Espacement entre le cadre du graphique et la l√©gende
)

# On ajuste automatiquement les marges pour que tout le contenu soit visible
plt.tight_layout()
# On affiche le graphique √† l'√©cran
plt.show()

```
